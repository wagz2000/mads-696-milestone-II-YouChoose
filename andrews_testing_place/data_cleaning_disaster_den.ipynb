```python
def get_transcripts(input_file):
    df = pd.read_csv(input_file)

    df['Transcripts_Raw_Json'] = ""
    df['Is_Generated'] = True
    for i, row in df.iterrows():
        print(f"{i}/{len(df.shape[0])}")
        try:
            transcripts = ytt.list_transcripts(row['Video ID'])

            transcripts = transcripts.find_transcript(['en'])
            row['Is_Generated'] = transcripts.is_generated
            transcript = str( transcripts.fetch() )
            print(transcript)
            row['Transcripts_Raw_Json'] = transcript

        except:
            pass
            

    
    return df
```

```python
video_transcripts = get_transcripts("data/videos-stats.csv")
```

```python
def get_transcript(video_id):
    try:
        transcripts = ytt.list_transcripts(video_id)
        transcripts = transcripts.find_transcript(['en'])
        is_generated = transcripts.is_generated
        raw_transcript = str(transcripts.fetch())
        return is_generated, raw_transcript

    except:
        print(f"{video_id} has no transcript")
    


def get_transcripts_better(input_file):
    df = pd.read_csv(input_file)
    df['Is_Generated'], df['Transcripts_Raw_Json'] = zip(*df['Video ID'].apply(get_transcript))
```

```python
video_transcripts = get_transcripts_better("data/videos-stats.csv")
```

```python
import pandas as pd
import string
import numpy as np
import nltk

# run this if first time
# nltk.download('punkt')
# nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
def cleaner(text, stop_words, stemmer):

    if np.isnan(text):
        return None
    try:
        text = text.lower()
        text = text.translate(str.maketrans('', '', string.punctuation))
        tokenized = word_tokenize(text)
        tokenized = [stemmer.stem(token) for token in tokenized if token not in stop_words]

        return ' '.join(tokenized)
    except Exception as e:
        print( e )
    
    return None

def clean_transcripts(input_file):
    stop_words = set(stopwords.words('english'))
    df = pd.read_csv(input_file)
    stemmer = PorterStemmer()
    df['Transcript_Clean'] = df['Transcript_Blob'].apply(cleaner, args=(stop_words, stemmer))
    df['Title_Clean'] = df['Title'].apply(cleaner, args=(stop_words, stemmer))


    return df

```

```python
video_transcripts = clean_transcripts('data/video_transcripts_blob.csv')
```
